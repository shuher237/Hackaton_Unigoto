```python
import numpy as np
import pandas as pd
import requests

from google.colab import drive
```

```python
drive.mount('/content/drive')
home_dir = "/content/drive/My Drive/Colab Notebooks/hackaton/"
token = '12345'
np.random.seed(seed=42)
```

```python
pages_list = np.random.randint(low=2, high=700000, size=250000)

first_response = requests.get(f'http://45.91.8.110/api/1?token={token}')
df = pd.DataFrame(first_response.json()['response'])

for page_number in pages_list:
    response = requests.get(f'http://45.91.8.110/api/{page_number}?token={token}')
    if response.status_code == 200:
        df_buf = pd.DataFrame(response.json()['response'])
        df = pd.concat([df, df_buf], ignore_index=True)

df.to_csv(home_dir+'data.csv', index=False)
```
